```python
# Import Statements
import pandas as pd
import folium
from folium import plugins
```

```python
# Load Data
breweries = pd.read_csv('Data/wa_breweries.csv')
```

```python
# Base Map
base_map = folium.Map(location=[47.7511, -120.7401], zoom_start=7)

# Add Markers for Breweries
for index, row in breweries.iterrows():
    brewery_name = row['brewery']
    
    # Clean latitude and longitude values
    brewery_lat = row['latitude']
    brewery_lon = row['longitude']
    
    brewery_popup = f"{brewery_name}"
    folium.Marker(location=[brewery_lat, brewery_lon], popup=brewery_popup).add_to(base_map)
    
# Display Map
base_map.save("Viz/breweries_map.html")
base_map
```
