```python
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
# Upload data
breweries = pd.read_csv('Data/wa_breweries.csv')
prod22 = pd.read_csv('Data/brewery_production_2022.csv')
prod1719 = pd.read_csv('Data/brewery_production_2017_2019.csv')
```

```python
# get unique names for each
brewery_names = pd.DataFrame(breweries['brewery'].unique(), columns =['brewery'])
brewery_names22 = pd.DataFrame(prod22['brewery_name'].unique(), columns =['brewery_22'])
brewery_names1719 = pd.DataFrame(prod1719['brewery_name'].unique(), columns =['brewery_1719'])
```

```python
df1 = brewery_names.merge(brewery_names22, how='outer', left_on='brewery', right_on='brewery_22')
```

```python
# breweries missing from breweries df
df1[df1['brewery'].isna()].head(50)
```

```python
# breweries missing from production data
df1[df1['brewery_22'].isna()].head(100)
```

```python
df2 = df1.merge(brewery_names1719, how='outer', left_on='brewery', right_on='brewery_1719')
```
